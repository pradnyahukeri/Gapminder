```python
def load_image(base_path):
    """it loads all the image into X and the classes in y """
    X_list = [] # makes a blank list for X data
    y_list = [] #same for y
    classes = os.listdir(base_path) #looks inside the base_path folder (in my example in  ./data/ )
    
    for class_ in classes:
        if class_ != ".DS_Store":
            files = os.listdir(base_path+class_)

            for file in files:
                if file != ".DS_Store": #loads everything but .DS_Store
                    pic = keras.preprocessing.image.load_img(path=base_path+class_+'/'+f'{file}',target_size=(224,224))
                    numpy_image = np.array(pic)
                
                #important! this is where we preprocess each image as we load it in using VGG16's preprocessing
                #(see imports)
                    processed_image = preprocess_input(numpy_image) 
            
                    X_list.append(processed_image)
                    y_list.append(class_)
        
    X = np.array(X_list)
    y = np.array(y_list)
    return X, y, classes
```

```python
X,y,classes= load_image(base_path)
```

## Base Model + Customized Dense Layers for Classification

+ fc1 with 30 neurons and activation relu
+ fc2 with 15 neurons and activation relu
+ prediction with 1 neurons for classification

```python
from tensorflow.keras import backend as K
```

```python
base_model.summary()
```

#### Configure the model

```python
K.clear_session()
new_model = keras.models.Sequential([
    
    # Base model
    base_model,
    
    # Flatten layer
    keras.layers.Flatten(),
    
    #keras.layers.Dropout(.5), #optional
    # fc1 fully connected layer
    keras.layers.Dense(units=30, activation=keras.activations.relu, name='fc1'),
    
    # fc2 fully connected layer
    keras.layers.Dense(units=15, activation=keras.activations.relu, name='fc2'),
    
    # output layer
    keras.layers.Dense(units=3, activation=keras.activations.softmax, name='output')

])
```

```python
new_model.summary()
```

#### Compile the new model

```python
loss_fn=keras.losses.SparseCategoricalCrossentropy(from_logits=False)
```

```python
new_model.compile(loss=loss_fn,
                  optimizer=keras.optimizers.Adam(learning_rate=1e-5),
                  metrics=['accuracy'])
```

#### Fit the new model
