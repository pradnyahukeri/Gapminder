```python
import pandas as pd
from prophet import Prophet
import numpy as np
```

```python
df=pd.read_csv("TG_STAID002759.txt",sep=',',header=14,parse_dates=[1])
df
```

```python
df.columns=['souid','date','temp','quality']
df
```

```python
df.drop(columns=['souid'],inplace=True)
df
```

```python
df['temp']=df['temp']/10
```

```python
df['temp'].describe()
```

```python
df['quality'].value_counts()
```

```python
df['temp'].replace({-999.9:np.nan},inplace=True)
```

```python
df
```

```python
# dropped na values
df.dropna(inplace=True)
```

```python
df
```

```python
df=df.drop('quality',axis=1)
```

```python
df
```

```python
m=Prophet()
```

```python
df.columns=['ds','y']
```

```python
df
```

```python
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=60)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(60)
```

```python
# Python
fig1 = m.plot(forecast)
```
