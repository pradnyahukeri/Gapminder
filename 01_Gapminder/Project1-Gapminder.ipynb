```python
for year in range(1960,2016):   
    df_subset = df_gapminder.loc[df_gapminder['year'] == year]
    sns.scatterplot(x='life expectancy', y='fertility_rate',size="total population",sizes=(10,1000),hue='continent',data=df_subset, alpha=0.6)
    plt.axis((0, 90, 0, 10))
    plt.legend(loc=2, bbox_to_anchor=(1, 1))
    plt.title(f'life expectancy v/s fertility_rate {year}')
    lgd = plt.legend(loc=2, bbox_to_anchor=(1, 1))
    plt.savefig(f'lifeexp_{year}.png', bbox_extra_artists=(lgd,), bbox_inches='tight')
    plt.close()
```
